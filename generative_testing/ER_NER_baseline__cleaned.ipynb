```python
def computeEmbeddings(text):
  embeddings = model.encode(text)
  return embeddings


```

```python
def computeEmbeddingsAllDocuments(train):
  data_emb=list()
  for entry in train:
    #print(id)
    #print(row)
    entry_emb=entry
    embeddings=computeEmbeddings(entry["text"])
    entry_emb["embeddings"]=embeddings
    data_emb.append(entry_emb)

  return data_emb


```

#Compute SimilarityMaps

```python
import pickle


train_embeddings=pickle.load(open(os.path.join(dataset_folder,"embeddings_train.p"), "rb"))
val_embeddings=pickle.load(open(os.path.join(dataset_folder,"embeddings_val.p"), "rb"))
test_embeddings=pickle.load(open(os.path.join(dataset_folder,"embeddings_test.p"), "rb"))
```

```python
from sentence_transformers import util

def findMostSimilarEmbedding(test_entry,train_embeddings):
  highest_similarity=0
  most_similar_id=-1
  for train_entry in train_embeddings:
    similarity=util.cos_sim(test_entry["embeddings"],train_entry["embeddings"]).item()
    #print(similarity)
    if similarity>highest_similarity:
      #print(similarity,highest_similarity)
      highest_similarity=similarity
      most_similar_id=train_entry["doc_id"]




  return highest_similarity,most_similar_id
```

```python
def precomputeSimilarities(test):
  similarity_map=list()
  for entry in test:
    temp=dict()
    temp["doc_id"]=entry["doc_id"]
    similarity,most_similar_id=findMostSimilarEmbedding(entry,train_embeddings)
    temp["train_most_similar_id"]=most_similar_id
    temp["similarity"]=similarity
    similarity_map.append(temp)

  return similarity_map



```

```python
similarity_map=precomputeSimilarities(test_embeddings)
```

```python
import pandas as pd
similarity_map_df=pd.DataFrame.from_dict(similarity_map)
```

```python
#Convert Polarity Label

def convertPolarity(dataset):
  new_dataset=list()

  for entry in dataset:
    new_entry=dict()
    new_entry["doc_id"]=entry["doc_id"]
    new_entry["text"]=entry["text"]
    new_entry["annotations"]=list()

    for annotation in entry["annotations"]:
      new_annotation=annotation.copy()
      if annotation["label"]=="Alergias medicamentosas":
        if annotation["Polaridade"]=="Positiva":
            new_annotation["label"]="Alergias medicamentosas positiva"
        else:
            new_annotation["label"]="Alergias medicamentosas negativa"
      else:
        new_annotation["label"]=annotation["label"]
      new_entry["annotations"].append(new_annotation)
    new_dataset.append(new_entry)
  return new_dataset
```

```python
def getTextSpanFromPosition(entry):
  text=entry["text"]
  annotations_res=list()
  for annotation in entry["annotations"]:
    temp=dict()
    temp["begin"]=annotation["begin"]
    temp["end"]=annotation["end"]
    temp["span"]=text[temp["begin"]:temp["end"]]
    temp["label"]=annotation["label"]
    annotations_res.append(temp)
  return annotations_res

import langextract as lx
def getExamples(doc_id,train_dataset):
  """
  Givena  doc_id from the training get the annotation from the training dataset
  """
  #print(doc_id)
  entry=[x for x in train_dataset if x["doc_id"]==doc_id][0]
  #print(extractions)
  annotations=getTextSpanFromPosition(entry)

  #print(extractions)
  lx_extract=list()
  for annotation in annotations:
    lx_extract.append(lx.data.Extraction(extraction_class=annotation["label"], extraction_text=annotation["span"]))


  #print(narrative,lang)
  # Define example data with entities in order of appearance
  examples = [
      lx.data.ExampleData(
          text=entry["text"],
          extractions=lx_extract
      )
  ]
  return examples
```

```python
id_ex=train[0]["doc_id"]
getTextSpanFromPosition(train[0])
getExamples(id_ex,train)
```

```python
def getDynamicExample(doc_id,train,similarity_map_df):
  matching=similarity_map_df[similarity_map_df["doc_id"]==doc_id].iloc[0]
  most_similar_id=matching["train_most_similar_id"]
  examples=getExamples(most_similar_id,train)
  return examples
```

```python
prompt='''
Extract structured clinical entities from emergency room admission notes.

Extract only information explicitly present in the text. Do not infer, normalize, or add entities that are not stated.

Entity classes to extract:

1. principal_diagnosis
- The main clinically specific diagnosis associated with the current admission or clinical episode.
- Extract the full diagnosis span as written. If several principal diagnosis appear listed, extract only the first.
- Do not extract vague background conditions, symptoms, procedures, or generic disease categories unless clearly presented as the principal diagnosis.

2. usual_medication
- The patient’s chronic, habitual, home, or usual medication.
- Extract each medication separately.
- The span should include the medication name and, when present, dose, formulation, route, frequency, timing, and administration instructions.
- Do not extract medications prescribed only for the current emergency episode unless they are explicitly described as usual/chronic medication.

3. drug_allergy
- Medication allergies and explicit absence of medication allergies.
- For positive allergy statements, extract the allergenic medication or drug class.
- For negative allergy statements, extract the span indicating absence of drug allergies. Do not include the negation marker in the span (e.g. "sem")
- Assign polarity:
  - positive: a specific drug allergy is present
  - negative: the note explicitly states no known drug allergies


General extraction rules:
- Preserve original wording, spelling, abbreviations, and accents.
- Extract minimal but complete spans.
- Do not merge separate medications or allergies into one entity.
- Return no entity when the information is absent.
- Do not extract non-drug allergies.
- Do not extract past medical history unless it is clearly the principal diagnosis for the current admission.
'''
```

```python
import langextract as lx
from langextract.providers.gemini import GeminiLanguageModel

LANGEXTRACT_API_KEY="PLACE API KEY"
# switch model to gemma-4-31b-it for the other baseline

def extract_info_gemini_models(prompt,example_annotations,input_text,model="gemini-2.5-flash-lite"):
  model = GeminiLanguageModel(
    model_id=model,
    api_key=LANGEXTRACT_API_KEY,   # or your actual key
  )
  # Define extraction prompt
  prompt_description = prompt

  result = lx.extract(
    text_or_documents=input_text,
    prompt_description=prompt_description,
    examples=example_annotations,
    api_key=LANGEXTRACT_API_KEY,
    model=model

)
  return result

```
